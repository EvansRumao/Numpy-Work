Regression Plots 

lmplot()
regression plot is used in machine learning so we will not fully discuss what, how and where it is used but we will see how it is created and what are the values we can put inside it

```python
tips = sns.load_dataset("tips")
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips)
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips,hue = 'sex')
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips,hue = 'sex',palette='coolwarm')
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips,hue = 'sex',palette='coolwarm',markers=['o','v'],
           scatter_kws={'s':50})
```

```python
import seaborn as sns
import pandas as pd 
import cufflinks as cf 
from plotly.offline import iplot
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
import cufflinks as cf
cf.go_offline()
cf.set_config_file(offline=True, world_readable=True)
```

```python
import plotly.express as px
px.line(tips, y='total_bill', title='Total Bill Trend')
```

```python
import plotly.express as px

df = tips.groupby('day', as_index=False)['tip'].mean()

px.bar(df, x='day', y='tip', title='Average Tip per Day')
```
