```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}
df = pd.DataFrame(data)
```

```python
df
```

# Group by Category and calculate the sum of Sales


```python
cat=df.groupby('Category')
for name, group in cat:
    print(f"Category: {name}")
    print(group)
    print("\n")
```

```python
cat['Sales'].sum() # Grouping by 'Category' and summing the 'Sales' column
```

```python
cat=df.groupby('Store')['Sales'].sum() # Grouping by 'Store' and summing the 'Sales' column
cat
```

```python
# Group by multiple columns
# Group by Category and Store
cat = df.groupby(['Category','Store'])['Sales'].sum()
cat
```

Aggregation

```python
df['Sales'].mean() # Calculate the mean sales for each category
```

```python
df['Sales'].agg(['sum', 'mean', 'min', 'max', 'count', 'std', 'median']) # Calculate multiple aggregate functions for the 'Sales' column
```

pivot Tables

```python
data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': ['East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West',
               'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': ['John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary',
            'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice']
}

df = pd.DataFrame(data)
```

```python
df
```

```python
pd.pivot_table(df,values='Sales',index='Product',columns='Region') # Create a pivot table to summarize sales by product and region  
```

```python
pd.pivot_table(df,values='Sales',index='Product',columns='Region',aggfunc='median') # Create a pivot table to summarize median sales by product and region
```

```python
pivot2 = pd.pivot_table(df, values=['Sales', 'Units'], index='Region', columns='Product')
pivot2
```

```python
pd.crosstab(df['Region'],df['Product']) # Create a crosstab to show the frequency of products sold in each region
```
