DataFrames using Pandas

```python
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}
df = pd.DataFrame(data)
```

```python
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
df2 = pd.DataFrame(data_list)
columns = ["Name","Age","City","Salary"]
df2 = pd.DataFrame(data_list,columns =columns)
df2
```

Selecting a Row

```python
df2.loc[[0,1]]
```

```python
df2.iloc[3] #iloc is used to access rows by their integer position, while loc is used to access rows by their label. In this case, df2.iloc[3] will return the fourth row of the DataFrame, which corresponds to the data for "Linda".
```

Selecting subsets of rows and coloumns

```python
df.loc[[1,2]][["City","Salary"]] #loc is used to access rows and columns by their labels. In this case, df.loc[[1,2]] will return the rows with labels 1 and 2, which correspond to "Anna" and "Peter". Then, [["City","Salary"]] is used to select only the "City" and "Salary" columns from those rows.
```

```python
df.loc[[2,3]][["Name","Age"]]
```
